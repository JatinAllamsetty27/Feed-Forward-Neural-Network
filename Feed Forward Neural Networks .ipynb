```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

A sigmoid function is a mathematical function having a characteristic "S"-shaped curve or sigmoid curve.

```python
def feed_forward(inputs, weights):
    # Add bias to inputs
    inputs = np.append(inputs, 1)

    # Calculate the net inputs for the hidden layer
    hidden_inputs = np.dot(weights[0], inputs)

    # Apply activation function to the hidden layer
    hidden_outputs = sigmoid(hidden_inputs)

    # Add bias to the hidden layer outputs
    hidden_outputs = np.append(hidden_outputs, 1)

    # Calculate the net inputs for the output layer
    output_inputs = np.dot(weights[1], hidden_outputs)

    # Apply activation function to the output layer
    output_outputs = sigmoid(output_inputs)

    return output_outputs
```

```python
input_values = np.array([0.5, 0.3])  # Input values
hidden_weights = np.array([[0.2, 0.4, 0.1], [0.1, 0.3, 0.5]])  # Hidden layer weights
output_weights = np.array([[0.3, 0.5, 0.2]])  # Output layer weights
```

```python
# Combine the weights into a list
weights = [hidden_weights, output_weights]
```

```python
# Perform feed-forward computation
output = feed_forward(input_values, weights)
```

```python
print("Output:", output)
```

In this example, the sigmoid function is used as the activation function for both the hidden layer and the output layer. The feed_forward function takes the input values and the weights as inputs and calculates the output of the neural network using the feed-forward algorithm. The final output is printed at the end.

Note that this implementation assumes a single hidden layer neural network with a bias term.
